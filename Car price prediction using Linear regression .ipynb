```python
#manual encoding
## converting caregorical data into numerical data 
car_data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)
#one hot encoding
car_data = pd.get_dummies(car_data, columns=['Seller_Type', 'Transmission'], drop_first=True)

```

```python
# darker value indicates low value of corelation and lighter value indicates high value of corelation .
## generated a heatmap to visualize correlations between numeric columns 

numeric_data = car_data.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 4))
sns.heatmap(numeric_data.corr(), annot=True,)
plt.title('Correlation between the columns')
plt.show()

```

```python
X = car_data.drop(['Car_Name','Selling_Price'], axis=1)
y = car_data['Selling_Price']
```

```python
print("Shape of X is: ",X.shape)
print("Shape of y is: ", y.shape)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print("X_test shape:", X_test.shape)
print("X_train shape:", X_train.shape)
print("y_test shape: ", y_test.shape)
print("y_train shape:", y_train.shape)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LinearRegression()
```

```python

model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
## imports three performance evaluation metrics for regression models:
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
## evaluates the performance of a model's predictions:
print("MAE: ", (metrics.mean_absolute_error(pred, y_test)))
print("MSE: ", (metrics.mean_squared_error(pred, y_test)))
print("R2 score: ", (metrics.r2_score(pred, y_test)))
```
